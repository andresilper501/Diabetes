```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
plt.style.use('classic')
```

```python
# Carrega o dataset de exemplo
df = pd.read_csv('/content/drive/MyDrive/diabetes.csv')
```

```python
# Verifica se temos dados faltantes
df.isnull().sum()
```

```python
#x = np.array(0,100)
```

```python
# Plota a distribuição de Diabetes em relação à idade
out = pd.cut(df['Age'], bins=[20, 30, 40, 50, 60, 70, 80, 90], include_lowest=True);
diabetes = df[df.Outcome == 1]
out2 = pd.cut(diabetes['Age'], bins=[20, 30, 40, 50, 60, 70, 80, 90], include_lowest=True);

ax = out.value_counts(sort=False)
plot1 = ax.plot.bar(x='Age', stacked=False, label='Healthy')
ax_diabetes = out2.value_counts(sort=False)
plot2 = ax_diabetes.plot.bar(x='Age', stacked=False, color='red', title='Diabetes x Age', label='Diabetes')
diabetes_percentage = (ax_diabetes / ax) * 100
plot3 = diabetes_percentage.plot(x='Age', label='Percentage')

ticks = plt.xticks(ticks=[0, 1, 2, 3, 4, 5, 6], labels=['20 to 30', '30 to 40', '40 to 50', '50 to 60', '60 to 70', '70 to 80', '80 to 90'])
plt.xticks(rotation='horizontal');
plt.ylabel('Number of people');
plt.xlabel('Age');
plt.legend()
```
